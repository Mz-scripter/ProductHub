```python
import pandas as pd
```

```python
df = pd.read_json('../data/raw/womens-clothing.json')
```

```python
df
```

```python
print(df[['price', 'rating', 'review_count']].isnull().sum())
```

```python
df_cleaned = df.dropna(subset=['price', 'rating', 'review_count'])
```

```python
df_cleaned
```

```python
df_cleaned
```

```python
df_cleaned.loc[:, 'price'] = df_cleaned['price'].str.replace('US $', '').str.replace('/ea', '').str.replace('GBP ', '').str.replace('AU $', '').str.replace('EUR ', '').str.replace('C $', '').str.replace(',', '')
```

```python
df_cleaned
```

```python
df_cleaned.loc[:, 'price'] = df_cleaned['price'].astype(float).astype(int)
```

```python
df_cleaned
```

```python
# Rename 'price' to 'price_in_dollars' for better clarity
df_cleaned = df_cleaned.copy()
df_cleaned.rename(columns={'price': 'price_in_dollars'}, inplace=True)
```

```python
df_cleaned.head()
```

```python
# Add new column for price_in_naira
usd_to_ngn_rate = 1551
df_cleaned['price_in_naira'] = df_cleaned['price_in_dollars'] * usd_to_ngn_rate
```

```python
df_cleaned.head()
```

```python
df_cleaned.rename(columns={
    'name': 'product_name',
    'rating': 'product_rating',
    'review_count': 'number_of_reviews',
    'image_url': 'product_image_url',
    'category': 'product_category',
    'price_in_dollars': 'price_usd',
    'price_in_naira': 'price_ngn'
}, inplace=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned['number_of_reviews'] = df_cleaned['number_of_reviews'].str.replace('This item (', '').str.replace(')', '').str.replace(',', '').astype(int)
```

```python
df_cleaned.head(2)
```

```python
# Assign unique id
import uuid
df_cleaned['product_id'] = [str(uuid.uuid4()) for _ in range(len(df_cleaned))]
```

```python
# Reorder the columns
column_order = ['product_id', 'product_name', 'product_rating', 'number_of_reviews', 'product_image_url', 'product_category', 'price_usd', 'price_ngn']

df_cleaned = df_cleaned[column_order]
```

```python
df_cleaned.head()
```

```python
# Save the cleaned data
df_cleaned.to_json('../data/cleaned/womens-clothing.json', orient='records', indent=4)
```
